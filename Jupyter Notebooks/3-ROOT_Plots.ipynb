Normalize the angle coordinates:

```python
muon_dataframe["1dtPrimitive.phiB"] = muon_dataframe["1dtPrimitive.phiB"]/512.
muon_dataframe["2dtPrimitive.phiB"] = muon_dataframe["2dtPrimitive.phiB"]/512.
muon_dataframe["3dtPrimitive.phiB"] = muon_dataframe["3dtPrimitive.phiB"]/512.
muon_dataframe["4dtPrimitive.phiB"] = muon_dataframe["4dtPrimitive.phiB"]/512.
```

```python
muon_dataframe.loc[muon_dataframe["1dtPrimitive.quality"] < 4, '1dtPrimitive.quality'] = 0.0
muon_dataframe.loc[muon_dataframe["1dtPrimitive.quality"] >= 4, '1dtPrimitive.quality'] = 1.0
muon_dataframe.loc[muon_dataframe["2dtPrimitive.quality"] < 4, '2dtPrimitive.quality'] = 0.0
muon_dataframe.loc[muon_dataframe["2dtPrimitive.quality"] >= 4, '2dtPrimitive.quality'] = 1.0
muon_dataframe.loc[muon_dataframe["3dtPrimitive.quality"] < 4, '3dtPrimitive.quality'] = 0.0
muon_dataframe.loc[muon_dataframe["3dtPrimitive.quality"] >= 4, '3dtPrimitive.quality'] = 1.0
muon_dataframe.loc[muon_dataframe["4dtPrimitive.quality"] < 4, '4dtPrimitive.quality'] = 0.0
muon_dataframe.loc[muon_dataframe["4dtPrimitive.quality"] >= 4, '4dtPrimitive.quality'] = 1.0
```

Preprocess the features in order to make prediction for the ROOT plot:

```python
def preprocess_features(muon_dataframe):
  """Prepares input features from Muon data set.

  Args:
    muon_dataframe: A Pandas DataFrame expected to contain data
      from muon simulations
  Returns:
    A DataFrame that contains the features to be used for the model.
  """
  selected_features = muon_dataframe[
[#'Event',
 'n_Primitive',
 '1dtPrimitive.id_r',
 '2dtPrimitive.id_r',
 '3dtPrimitive.id_r',
 '4dtPrimitive.id_r',
 '1dtPrimitive.id_eta',
 '3dtPrimitive.id_eta',
 '3dtPrimitive.id_eta',
 '4dtPrimitive.id_eta',
 '1dtPrimitive.id_phi',
 '2dtPrimitive.id_phi',
 '3dtPrimitive.id_phi',
 '4dtPrimitive.id_phi',
 '1dtPrimitive.phiB',
 '2dtPrimitive.phiB',
 '3dtPrimitive.phiB',
 '4dtPrimitive.phiB',
 '1dtPrimitive.quality',
 '2dtPrimitive.quality',
 '3dtPrimitive.quality',
 '4dtPrimitive.quality',
 'delta_phi12',
 'delta_phi13',
 'delta_phi14',
 'delta_phi23',
 'delta_phi24',
 'delta_phi34'
  ]]
  processed_features = selected_features.copy()
  return processed_features.astype(np.float32)
```

```python
outputFile.cd()
histos = {}
model = load_model('./model/my_model_mu4.h5')
bar = progressbar.ProgressBar(maxval=len(muon_dataframe.index), widgets = [progressbar.Bar('=','[',']'), '', progressbar.Percentage()])
count=0
all_count=0
count2=0
all_count2=0
count3=0
all_count3=0
count4=0
all_count4=0
#Plot for L1 vs GEN
histos["L1_SingleMuon22_Eff_Vs_Pt"] = root.TEfficiency("L1_SingleMuon22_Eff_Vs_Pt",   \
                                                       ";gen muon p_{T}; efficiency", \
                                                       100, 0., 200.)
histos["L1_SingleMuon27_Eff_Vs_Pt"] = root.TEfficiency("L1_SingleMuon27_Eff_Vs_Pt",   \
                                                       ";gen muon p_{T}; efficiency", \
                                                       100, 0., 200.)
histos["L1_SingleMuon32_Eff_Vs_Pt"] = root.TEfficiency("L1_SingleMuon32_Eff_Vs_Pt",   \
                                                       ";gen muon p_{T}; efficiency", \
                                                       100, 0., 200.)
histos["L1_SingleMuon80_Eff_Vs_Pt"] = root.TEfficiency("L1_SingleMuon80_Eff_Vs_Pt",   \
                                                       ";gen muon p_{T}; efficiency", \
                                                       100, 0., 200.)
histos["L1_SingleMuon100_Eff_Vs_Pt"] = root.TEfficiency("L1_SingleMuon100_Eff_Vs_Pt",   \
                                                       ";gen muon p_{T}; efficiency", \
                                                       100, 0., 200.)
histos["L1_SingleMuon140_Eff_Vs_Pt"] = root.TEfficiency("L1_SingleMuon140_Eff_Vs_Pt",   \
                                                       ";gen muon p_{T}; efficiency", \
                                                       100, 0., 200.)

histos["L1-GEN_qOverPt_Resol"] = root.TH1F("L1-GEN_qOverPt_Resol",   \
                                           ";q/p_{T} resolution; # entries", \
                                           200, -10., 10.)

histos["L1-GEN_oneOverPt_Resol"] = root.TH1F("L1-GEN_oneOverPt_Resol",   \
                                             ";1/p_{T} resolution; # entries", \
                                             200, -10., 10.)


histos["L1-GEN_Pt_Resol"] = root.TH1F("L1-GEN Pt Resol.",   \
                                      ";p_{T} resolution; # entries", \
                                      200, -10., 10.)
histos["L1-GEN_Pt_Resol1"] = root.TH1F("L1-GEN Pt Resol. 0-20GeV",   \
                                      "p_{T} resolution 0<p_{T}<20GeV/c;\Delta p_{T}/p_{T}; # entries", \
                                      200, -10., 10.)
histos["L1-GEN_Pt_Resol2"] = root.TH1F("L1-GEN Pt Resol. 20-40GeV",   \
                                      "p_{T} resolution 20<p_{T}<40GeV/c;\Delta p_{T}/p_{T}; # entries", \
                                      200, -10., 10.)
histos["L1-GEN_Pt_Resol3"] = root.TH1F("L1-GEN Pt Resol. 40-60GeV",   \
                                      "p_{T} resolution 40<p_{T}<60GeV/c;\Delta p_{T}/p_{T}; # entries", \
                                      200, -10., 10.)
histos["L1-GEN_Pt_Resol4"] = root.TH1F("L1-GEN Pt Resol. 60-80GeV",   \
                                      "p_{T} resolution 60<p_{T}<80GeV/c;\Delta p_{T}/p_{T}; # entries", \
                                      200, -10., 10.)
histos["L1-GEN_Pt_Resol5"] = root.TH1F("L1-GEN Pt Resol. 80-100GeV",   \
                                      "p_{T} resolution 80<p_{T}<100GeV/c;\Delta p_{T}/p_{T}; # entries", \
                                      200, -10., 10.)
histos["L1-GEN_Pt_Resol6"] = root.TH1F("L1-GEN_Pt_Resol100-150",   \
                                      "p_{T} resolution 100<p_{T}<150GeV/c;\Delta p_{T}/p_{T}; # entries", \
                                      200, -10., 10.)
histos["L1-GEN_Pt_Resol7"] = root.TH1F("L1-GEN_Pt_Resol150-200",   \
                                      "p_{T} resolution 150<p_{T}<200GeV/c;\Delta p_{T}/p_{T}; # entries", \
                                      200, -10., 10.)
histos["L1-GEN_Pt_Resol8"] = root.TH1F("L1-GEN_Pt_Resol0-120",   \
                                      "p_{T} resolution 3<p_{T}<120GeV/c;\Delta p_{T}/p_{T}; # entries", \
                                      200, -10., 10.)


#Plot for PRED_PT vs GEN
histos["PRED_SingleMuon22_Eff_Vs_Pt"] = root.TEfficiency("PRED_SingleMuon22_Eff_Vs_Pt",   \
                                                       ";gen muon p_{T}; efficiency", \
                                                       100, 0., 200.)
histos["PRED_SingleMuon27_Eff_Vs_Pt"] = root.TEfficiency("PRED_SingleMuon27_Eff_Vs_Pt",   \
                                                       ";gen muon p_{T}; efficiency", \
                                                       100, 0., 200.)
histos["PRED_SingleMuon32_Eff_Vs_Pt"] = root.TEfficiency("PRED_SingleMuon32_Eff_Vs_Pt",   \
                                                       ";gen muon p_{T}; efficiency", \
                                                       100, 0., 200.)
histos["PRED_SingleMuon80_Eff_Vs_Pt"] = root.TEfficiency("PRED_SingleMuon80_Eff_Vs_Pt",   \
                                                       ";gen muon p_{T}; efficiency", \
                                                       200, 0., 200.)
histos["PRED_SingleMuon100_Eff_Vs_Pt"] = root.TEfficiency("PRED_SingleMuon100_Eff_Vs_Pt",   \
                                                       ";gen muon p_{T}; efficiency", \
                                                       200, 0., 200.)
histos["PRED_SingleMuon140_Eff_Vs_Pt"] = root.TEfficiency("PRED_SingleMuon140_Eff_Vs_Pt",   \
                                                       ";gen muon p_{T}; efficiency", \
                                                       200, 0., 200.)

histos["PRED-GEN_qOverPt_Resol"] = root.TH1F("PRED-GEN_qOverPt_Resol",   \
                                           ";q/p_{T} resolution; # entries", \
                                           200, -10., 10.)

histos["PRED-GEN_oneOverPt_Resol"] = root.TH1F("PRED-GEN_oneOverPt_Resol",   \
                                             ";1/p_{T} resolution; # entries", \
                                             200, -10., 10.)


histos["PRED-GEN_Pt_Resol"] = root.TH1F("PRED-GEN Pt Resol.",   \
                                      ";p_{T} resolution; # entries", \
                                      200, -10., 10.)

histos["PRED-GEN_Pt_Resol1"] = root.TH1F("PRED-GEN Pt Resol. 0-20GeV",   \
                                      "p_{T} resolution 0<p_{T}<20GeV/c; \Delta p_{T}/p_{T}; # entries", \
                                      200, -10., 10.)
histos["PRED-GEN_Pt_Resol2"] = root.TH1F("PRED-GEN Pt Resol. 20-40GeV",   \
                                      "p_{T} resolution 20<p_{T}<40GeV/c; \Delta p_{T}/p_{T}; # entries", \
                                      200, -10., 10.)
histos["PRED-GEN_Pt_Resol3"] = root.TH1F("PRED-GEN Pt Resol. 40-60GeV",   \
                                      "p_{T} resolution 40<p_{T}<60GeV/c; \Delta p_{T}/p_{T}; # entries", \
                                      200, -10., 10.)
histos["PRED-GEN_Pt_Resol4"] = root.TH1F("PRED-GEN Pt Resol. 60-80GeV",   \
                                      "p_{T} resolution 60<p_{T}<80GeV/c; \Delta p_{T}/p_{T}; # entries", \
                                      200, -10., 10.)
histos["PRED-GEN_Pt_Resol5"] = root.TH1F("PRED-GEN Pt Resol. 80-100GeV",   \
                                      "p_{T} resolution 80<p_{T}<100GeV/c; \Delta p_{T}/p_{T}; # entries", \
                                      200, -10., 10.)
histos["PRED-GEN_Pt_Resol6"] = root.TH1F("PRED-GEN_Pt_Resol100-150",   \
                                      "p_{T} resolution 100<p_{T}<150GeV/c; \Delta p_{T}/p_{T}; # entries", \
                                      200, -10., 10.)
histos["PRED-GEN_Pt_Resol7"] = root.TH1F("PRED-GEN_Pt_Resol150-200",   \
                                      "p_{T} resolution 150<p_{T}<200GeV/c; \Delta p_{T}/p_{T}; # entries", \
                                      200, -10., 10.)
histos["PRED-GEN_Pt_Resol8"] = root.TH1F("PRED-GEN_Pt_Resol0-120",   \
                                      "p_{T} resolution 3<p_{T}<120GeV/c; \Delta p_{T}/p_{T}; # entries", \
                                      200, -10., 10.)
bar.start() 
for index, row in muon_dataframe.iterrows():  
    genMuonVector = root.TLorentzVector()
    genMuonVector.SetPtEtaPhiM(row['genParticle.pt'],  \
                                row['genParticle.eta'], \
                                row['genParticle.phi'], \
                                0.106)
    
    minDr = 999.
    minDr2 = 999.
    bestL1MuonVector = root.TLorentzVector()
    bestL1MuonVector2 = root.TLorentzVector()
    bestL1MuonCharge = 0
    bestL1MuonCharge2 = 0

    X = preprocess_features(muon_dataframe.iloc[[index]])
    pt_predict = model.predict(X) *200

    for i in range(1,5):

        # Looking just at SingleMuon qualities
        if row[str(i)+'l1Muon.quality'] < 12 :
            continue

        l1MuonVector = root.TLorentzVector()
        l1MuonVector.SetPtEtaPhiM(pt_predict,  \
                                    row[str(i)+'l1Muon.eta'], \
                                    row[str(i)+'l1Muon.phi'], \
                                    0.106)
        l1MuonVector2 = root.TLorentzVector()
        l1MuonVector2.SetPtEtaPhiM(row[str(i)+'l1Muon.pt'], \
                                  row[str(i)+'l1Muon.eta'], \
                                  row[str(i)+'l1Muon.phi'], \
                                  0.106)
            
        dR = genMuonVector.DeltaR(l1MuonVector)
        dR2 = genMuonVector.DeltaR(l1MuonVector2)
        
        if (dR < minDr):
            minDr = dR
            bestL1MuonVector = l1MuonVector
            bestL1MuonCharge = row[str(i)+'l1Muon.charge']
        if (dR2 < minDr2):
            minDr2 = dR2
            bestL1MuonVector2 = l1MuonVector2
            bestL1MuonCharge2 = row[str(i)+'l1Muon.charge']
            
    # We are only interested in pT assignment
    # therefore we require the dR matching to be < 0.3 already
    if minDr < 0.3 :
        if (genMuonVector.Pt()>=5 and genMuonVector.Pt()<=15):
            all_count2 = all_count2 + 1
            if (bestL1MuonVector.Pt()>=20):
                count2 = count2 + 1
                
        if (genMuonVector.Pt()>=30 and genMuonVector.Pt()<=190):
            all_count4 = all_count4 + 1
            if (bestL1MuonVector.Pt()>=20):
                count4 = count4 + 1

        histos["PRED_SingleMuon22_Eff_Vs_Pt"].Fill(bestL1MuonVector.Pt() >= 20, \
                                                 genMuonVector.Pt())
        
        histos["PRED_SingleMuon27_Eff_Vs_Pt"].Fill(bestL1MuonVector.Pt() >= 26, \
                                                 genMuonVector.Pt())
        
        histos["PRED_SingleMuon32_Eff_Vs_Pt"].Fill(bestL1MuonVector.Pt() >= 31, \
                                                 genMuonVector.Pt())
        
        histos["PRED_SingleMuon80_Eff_Vs_Pt"].Fill(bestL1MuonVector.Pt() >= 73, \
                                                 genMuonVector.Pt())
        
        histos["PRED_SingleMuon100_Eff_Vs_Pt"].Fill(bestL1MuonVector.Pt() >= 92, \
                                                 genMuonVector.Pt())
        
        histos["PRED_SingleMuon140_Eff_Vs_Pt"].Fill(bestL1MuonVector.Pt() >= 140, \
                                                 genMuonVector.Pt())

        ptResol = (bestL1MuonVector.Pt() - genMuonVector.Pt()) / genMuonVector.Pt()
        if 0 < genMuonVector.Pt() <= 20:
            ptResol_1 = (bestL1MuonVector.Pt() - genMuonVector.Pt()) / genMuonVector.Pt()
            histos["PRED-GEN_Pt_Resol1"].Fill(ptResol_1)

        if 20 < genMuonVector.Pt() <= 40:
            ptResol_2 = (bestL1MuonVector.Pt() - genMuonVector.Pt()) / genMuonVector.Pt()
            histos["PRED-GEN_Pt_Resol2"].Fill(ptResol_2)
            
        if 40 < genMuonVector.Pt() <= 60:
            ptResol_3 = (bestL1MuonVector.Pt() - genMuonVector.Pt()) / genMuonVector.Pt()
            histos["PRED-GEN_Pt_Resol3"].Fill(ptResol_3)

        if 60 < genMuonVector.Pt() <= 80:
            ptResol_4 = (bestL1MuonVector.Pt() - genMuonVector.Pt()) / genMuonVector.Pt()
            histos["PRED-GEN_Pt_Resol4"].Fill(ptResol_4)

        if 80 < genMuonVector.Pt() <= 100:
            ptResol_5 = (bestL1MuonVector.Pt() - genMuonVector.Pt()) / genMuonVector.Pt()
            histos["PRED-GEN_Pt_Resol5"].Fill(ptResol_5)
            
        if 100 < genMuonVector.Pt() <= 150:
            ptResol_6 = (bestL1MuonVector.Pt() - genMuonVector.Pt()) / genMuonVector.Pt()
            histos["PRED-GEN_Pt_Resol6"].Fill(ptResol_6)
        
        if 150 < genMuonVector.Pt() <= 200:
            ptResol_7 = (bestL1MuonVector.Pt() - genMuonVector.Pt()) / genMuonVector.Pt()
            histos["PRED-GEN_Pt_Resol7"].Fill(ptResol_7)
        
        if 0 < genMuonVector.Pt() <= 120:
            ptResol_8 = (bestL1MuonVector.Pt() - genMuonVector.Pt()) / genMuonVector.Pt()
            histos["PRED-GEN_Pt_Resol8"].Fill(ptResol_8)
            
        oneOverPtResol = (1. / bestL1MuonVector.Pt() - 1. / genMuonVector.Pt()) / (1. /genMuonVector.Pt())

        histos["PRED-GEN_Pt_Resol"].Fill(ptResol)
        histos["PRED-GEN_oneOverPt_Resol"].Fill(oneOverPtResol)

        if bestL1MuonCharge != 0 :
            qOverPtL1  = bestL1MuonCharge   / bestL1MuonVector.Pt() 
            qOverPtGEN = - abs(row['genParticle.pdgId']) / row['genParticle.pdgId']  / genMuonVector.Pt() 
            qOverPtResol = (qOverPtL1 - qOverPtGEN) / qOverPtGEN

            histos["PRED-GEN_qOverPt_Resol"].Fill(qOverPtResol)
    
    if minDr2 < 0.3 :
        if (genMuonVector.Pt()>=5 and genMuonVector.Pt()<=15):
            all_count = all_count + 1
            if (bestL1MuonVector2.Pt()>=22):
                count = count + 1
                
        if (genMuonVector.Pt()>=30 and genMuonVector.Pt()<=190):
            all_count3 = all_count3 + 1
            if (bestL1MuonVector2.Pt()>=22):
                count3 = count3 + 1
                
        histos["L1_SingleMuon22_Eff_Vs_Pt"].Fill(bestL1MuonVector2.Pt() >= 22, \
                                                 genMuonVector.Pt())
        
        histos["L1_SingleMuon27_Eff_Vs_Pt"].Fill(bestL1MuonVector2.Pt() >= 27, \
                                                 genMuonVector.Pt())
        
        histos["L1_SingleMuon32_Eff_Vs_Pt"].Fill(bestL1MuonVector2.Pt() >= 32, \
                                                 genMuonVector.Pt())
        
        histos["L1_SingleMuon80_Eff_Vs_Pt"].Fill(bestL1MuonVector2.Pt() >= 80, \
                                                 genMuonVector.Pt())
        
        histos["L1_SingleMuon100_Eff_Vs_Pt"].Fill(bestL1MuonVector2.Pt() >= 100, \
                                                 genMuonVector.Pt())
        
        histos["L1_SingleMuon140_Eff_Vs_Pt"].Fill(bestL1MuonVector2.Pt() >= 140, \
                                                 genMuonVector.Pt())

        ptResol2 = (bestL1MuonVector2.Pt() - genMuonVector.Pt()) / genMuonVector.Pt()
        if 0 < genMuonVector.Pt() <= 20:
            ptResol2_1 = (bestL1MuonVector2.Pt() - genMuonVector.Pt()) / genMuonVector.Pt()
            histos["L1-GEN_Pt_Resol1"].Fill(ptResol2_1)
            
        if 20 < genMuonVector.Pt() <= 40:
            ptResol2_2 = (bestL1MuonVector2.Pt() - genMuonVector.Pt()) / genMuonVector.Pt()
            histos["L1-GEN_Pt_Resol2"].Fill(ptResol2_2)
            
        if 40 < genMuonVector.Pt() <= 60:
            ptResol2_3 = (bestL1MuonVector2.Pt() - genMuonVector.Pt()) / genMuonVector.Pt()
            histos["L1-GEN_Pt_Resol3"].Fill(ptResol2_3)
            
        if 60 < genMuonVector.Pt() <= 80:
            ptResol2_4 = (bestL1MuonVector2.Pt() - genMuonVector.Pt()) / genMuonVector.Pt()
            histos["L1-GEN_Pt_Resol4"].Fill(ptResol2_4)
            
        if 80 < genMuonVector.Pt() <= 100:
            ptResol2_5 = (bestL1MuonVector2.Pt() - genMuonVector.Pt()) / genMuonVector.Pt()
            histos["L1-GEN_Pt_Resol5"].Fill(ptResol2_5)
            
        if 100 < genMuonVector.Pt() <= 150:
            ptResol2_6 = (bestL1MuonVector2.Pt() - genMuonVector.Pt()) / genMuonVector.Pt()
            histos["L1-GEN_Pt_Resol6"].Fill(ptResol2_6)
            
        if 150 < genMuonVector.Pt() <= 200:
            ptResol2_7 = (bestL1MuonVector2.Pt() - genMuonVector.Pt()) / genMuonVector.Pt()
            histos["L1-GEN_Pt_Resol7"].Fill(ptResol2_7)
        
        if 0 < genMuonVector.Pt() <= 120:
            ptResol2_8 = (bestL1MuonVector2.Pt() - genMuonVector.Pt()) / genMuonVector.Pt()
            histos["L1-GEN_Pt_Resol8"].Fill(ptResol2_8)

        oneOverPtResol2 = (1. / bestL1MuonVector2.Pt() - 1. / genMuonVector.Pt()) / (1. /genMuonVector.Pt())

        histos["L1-GEN_Pt_Resol"].Fill(ptResol2)
        histos["L1-GEN_oneOverPt_Resol"].Fill(oneOverPtResol2)

        if bestL1MuonCharge2 != 0 :
            qOverPtL1_2  = bestL1MuonCharge2   / bestL1MuonVector2.Pt() 
            qOverPtGEN_2 = - abs(row['genParticle.pdgId']) / row['genParticle.pdgId']  / genMuonVector.Pt() 
            qOverPtResol_2 = (qOverPtL1_2 - qOverPtGEN_2) / qOverPtGEN_2

            histos["L1-GEN_qOverPt_Resol"].Fill(qOverPtResol_2)
    bar.update(index+1)
c1 = root.TCanvas("c1","c1"); 
histos["PRED_SingleMuon22_Eff_Vs_Pt"].SetLineColor(kBlue);
#histos["PRED_SingleMuon22_Eff_Vs_Pt"].SetMarkerStyle(20);
histos["PRED_SingleMuon22_Eff_Vs_Pt"].SetMarkerColor(kBlue);
#histos["PRED_SingleMuon22_Eff_Vs_Pt"].SetMarkerSize(2.3);
histos["L1_SingleMuon22_Eff_Vs_Pt"].SetLineColor(kRed); 
#histos["L1_SingleMuon22_Eff_Vs_Pt"].SetMarkerStyle(22);
histos["L1_SingleMuon22_Eff_Vs_Pt"].SetMarkerColor(kRed);
#histos["L1_SingleMuon22_Eff_Vs_Pt"].SetMarkerSize(2.4);
histos["L1_SingleMuon22_Eff_Vs_Pt"].SetTitle("Predicted and Level-1 Efficiency curves;Muon P_{T} (GeV/c);Efficiency(%)");
histos["L1_SingleMuon22_Eff_Vs_Pt"].Draw();
histos["PRED_SingleMuon22_Eff_Vs_Pt"].Draw("SAMES");
c1.Update();
leg = root.TLegend(0.32,0.5,0.90,0.67);
leg.SetTextSize(0.035);
leg.SetHeader("Efficiency:"); 
leg.AddEntry(histos["PRED_SingleMuon22_Eff_Vs_Pt"], "NEW: Machine Learning  @22GeV/c", "lep");
leg.AddEntry(histos["L1_SingleMuon22_Eff_Vs_Pt"], "CURRENT: Level-1 Trigger @22GeV/c", "lep");
leg.Draw();
c1.Write("Turnon Curves 22");

c2 = root.TCanvas("c2","c2"); 
histos["PRED_SingleMuon27_Eff_Vs_Pt"].SetLineColor(kBlue);
#histos["PRED_SingleMuon40_Eff_Vs_Pt"].SetMarkerStyle(20);
histos["PRED_SingleMuon27_Eff_Vs_Pt"].SetMarkerColor(kBlue);
#histos["PRED_SingleMuon40_Eff_Vs_Pt"].SetMarkerSize(2.3);
histos["L1_SingleMuon27_Eff_Vs_Pt"].SetLineColor(kRed); 
#histos["L1_SingleMuon40_Eff_Vs_Pt"].SetMarkerStyle(22);
histos["L1_SingleMuon27_Eff_Vs_Pt"].SetMarkerColor(kRed);
#histos["L1_SingleMuon40_Eff_Vs_Pt"].SetMarkerSize(2.4);
histos["L1_SingleMuon27_Eff_Vs_Pt"].SetTitle("Predicted and Level-1 Efficiency curves;Muon P_{T} (GeV/c);Efficiency(%)");
histos["L1_SingleMuon27_Eff_Vs_Pt"].Draw();
histos["PRED_SingleMuon27_Eff_Vs_Pt"].Draw("SAMES");
c2.Update();
leg2 = root.TLegend(0.49,0.5,0.90,0.63);
leg2.SetTextSize(0.035);
leg2.SetHeader("Efficiency:"); 
leg2.AddEntry(histos["PRED_SingleMuon27_Eff_Vs_Pt"], "NEW: Machine Learning  @27GeV/c", "lep");
leg2.AddEntry(histos["L1_SingleMuon27_Eff_Vs_Pt"], "CURRENT: Level-1 Trigger @27GeV/c", "lep");
leg2.Draw();
c2.Write("Turnon Curves 27");

c3 = root.TCanvas("c3","c3"); 
histos["PRED_SingleMuon32_Eff_Vs_Pt"].SetLineColor(kBlue);
#histos["PRED_SingleMuon60_Eff_Vs_Pt"].SetMarkerStyle(20);
histos["PRED_SingleMuon32_Eff_Vs_Pt"].SetMarkerColor(kBlue);
#histos["PRED_SingleMuon60_Eff_Vs_Pt"].SetMarkerSize(2.3);
histos["L1_SingleMuon32_Eff_Vs_Pt"].SetLineColor(kRed); 
#histos["L1_SingleMuon60_Eff_Vs_Pt"].SetMarkerStyle(22);
histos["L1_SingleMuon32_Eff_Vs_Pt"].SetMarkerColor(kRed);
#histos["L1_SingleMuon60_Eff_Vs_Pt"].SetMarkerSize(2.4);
histos["L1_SingleMuon32_Eff_Vs_Pt"].SetTitle("Predicted and Level-1 Efficiency curves;Muon P_{T} (GeV/c);Efficiency(%)");
histos["L1_SingleMuon32_Eff_Vs_Pt"].Draw();
histos["PRED_SingleMuon32_Eff_Vs_Pt"].Draw("SAMES");
c3.Update();
leg3 = root.TLegend(0.32,0.5,0.90,0.67);
leg3.SetTextSize(0.035);
leg3.SetHeader("Efficiency:"); 
leg3.AddEntry(histos["PRED_SingleMuon32_Eff_Vs_Pt"], "NEW: Machine Learning  @32GeV/c", "lep");
leg3.AddEntry(histos["L1_SingleMuon32_Eff_Vs_Pt"], "CURRENT: Level-1 Trigger @32GeV/c", "lep");
leg3.Draw();
c3.Write("Turnon Curves 32");

c4 = root.TCanvas("c4","c4"); 
histos["PRED_SingleMuon80_Eff_Vs_Pt"].SetLineColor(kBlue);
#histos["PRED_SingleMuon80_Eff_Vs_Pt"].SetMarkerStyle(20);
histos["PRED_SingleMuon80_Eff_Vs_Pt"].SetMarkerColor(kBlue);
#histos["PRED_SingleMuon80_Eff_Vs_Pt"].SetMarkerSize(2.3);
histos["L1_SingleMuon80_Eff_Vs_Pt"].SetLineColor(kRed);
#histos["L1_SingleMuon80_Eff_Vs_Pt"].SetMarkerStyle(22);
histos["L1_SingleMuon80_Eff_Vs_Pt"].SetMarkerColor(kRed);
#histos["L1_SingleMuon80_Eff_Vs_Pt"].SetMarkerSize(2.4);
histos["L1_SingleMuon80_Eff_Vs_Pt"].SetTitle("Predicted and Level-1 Efficiency curves;Muon P_{T} (GeV/c);Efficiency(%)");
histos["L1_SingleMuon80_Eff_Vs_Pt"].Draw();
histos["PRED_SingleMuon80_Eff_Vs_Pt"].Draw("SAMES");
c4.Update();
leg4 = root.TLegend(0.49,0.5,0.90,0.63);
leg4.SetTextSize(0.035);
leg4.SetHeader("Legend:"); 
leg4.AddEntry(histos["PRED_SingleMuon80_Eff_Vs_Pt"], "NEW: Machine Learning  @80GeV/c", "lep");
leg4.AddEntry(histos["L1_SingleMuon80_Eff_Vs_Pt"], "CURRENT: Level-1 Trigger @80GeV/c", "lep");
leg4.Draw();
c4.Write("Turnon Curves 80");

c5 = root.TCanvas("c5","c5"); 
histos["PRED_SingleMuon100_Eff_Vs_Pt"].SetLineColor(kBlue);
#histos["PRED_SingleMuon100_Eff_Vs_Pt"].SetMarkerStyle(20);
histos["PRED_SingleMuon100_Eff_Vs_Pt"].SetMarkerColor(kBlue);
#histos["PRED_SingleMuon100_Eff_Vs_Pt"].SetMarkerSize(2.3);
histos["L1_SingleMuon100_Eff_Vs_Pt"].SetLineColor(kRed); 
#histos["L1_SingleMuon100_Eff_Vs_Pt"].SetMarkerStyle(22);
histos["L1_SingleMuon100_Eff_Vs_Pt"].SetMarkerColor(kRed);
#histos["L1_SingleMuon100_Eff_Vs_Pt"].SetMarkerSize(2.4);
histos["L1_SingleMuon100_Eff_Vs_Pt"].SetTitle("Predicted and Level-1 Efficiency curves;Muon P_{T} (GeV/c);Efficiency(%)");
histos["L1_SingleMuon100_Eff_Vs_Pt"].Draw();
histos["PRED_SingleMuon100_Eff_Vs_Pt"].Draw("SAMES");
c5.Update();
leg5 = root.TLegend(0.32,0.5,0.90,0.67);
leg5.SetTextSize(0.035);
leg5.SetHeader("Legend:"); 
leg5.AddEntry(histos["PRED_SingleMuon100_Eff_Vs_Pt"], "NEW: Machine Learning @100GeV/c", "lep");
leg5.AddEntry(histos["L1_SingleMuon100_Eff_Vs_Pt"], "CURRENT: Level-1 Trigger @100GeV/c", "lep");
leg5.Draw();
c5.Write("Turnon Curves 100");

c6 = root.TCanvas("c6","c6"); 
histos["PRED_SingleMuon140_Eff_Vs_Pt"].SetLineColor(kBlue);
#histos["PRED_SingleMuon140_Eff_Vs_Pt"].SetMarkerStyle(20);
histos["PRED_SingleMuon140_Eff_Vs_Pt"].SetMarkerColor(kBlue);
#histos["PRED_SingleMuon140_Eff_Vs_Pt"].SetMarkerSize(2.3);
histos["L1_SingleMuon140_Eff_Vs_Pt"].SetLineColor(kRed); 
#histos["L1_SingleMuon140_Eff_Vs_Pt"].SetMarkerStyle(22);
histos["L1_SingleMuon140_Eff_Vs_Pt"].SetMarkerColor(kRed);
#histos["L1_SingleMuon140_Eff_Vs_Pt"].SetMarkerSize(2.4);
histos["PRED_SingleMuon140_Eff_Vs_Pt"].SetTitle("Predicted and Level-1 Efficiency curves;Muon P_{T} (GeV/c);Efficiency(%)");
histos["PRED_SingleMuon140_Eff_Vs_Pt"].Draw();
histos["L1_SingleMuon140_Eff_Vs_Pt"].Draw("SAMES");
c6.Update();
leg6 = root.TLegend(0.49,0.5,0.90,0.63);
leg6.SetTextSize(0.035);
leg6.SetHeader("Legend:"); 
leg6.AddEntry(histos["PRED_SingleMuon140_Eff_Vs_Pt"], "PREDICTED Eff. vs P_{T}@140GeV/c", "lep");
leg6.AddEntry(histos["L1_SingleMuon140_Eff_Vs_Pt"], "L1 Eff. vs P_{T}@140GeV/c", "lep");
leg6.Draw();
c6.Write("Turnon Curves 140");

c7 = root.TCanvas("c7","c7");
histos["L1-GEN_Pt_Resol"].SetLineColor(kRed)
histos["PRED-GEN_Pt_Resol"].SetTitle("p_{T} resolution comparison")
histos["PRED-GEN_Pt_Resol"].SetLineColor(kBlue)
histos["PRED-GEN_Pt_Resol"].Draw()
histos["L1-GEN_Pt_Resol"].Draw("SAMES")
st = root.TPaveStats()
st.FindObject("stats")
c7.Update()
leg7 = root.TLegend(0.49,0.5,0.90,0.63);
leg7.SetTextSize(0.04);
leg7.SetHeader("p_{T} Resolution:"); 
leg7.AddEntry(histos["PRED-GEN_Pt_Resol"], "NEW: Machine Learning", "l");
leg7.AddEntry(histos["L1-GEN_Pt_Resol"], "CURRENT: Level-1 trigger", "l");
leg7.Draw();
c7.Write("Pt-resol")

c8 = root.TCanvas("c8","c8");
histos["L1-GEN_Pt_Resol1"].SetLineColor(kRed)
histos["L1-GEN_Pt_Resol1"].SetTitle("p_{T} resolution comparison - Range 3-20 GeV")
histos["PRED-GEN_Pt_Resol1"].SetLineColor(kBlue)
histos["L1-GEN_Pt_Resol1"].Draw()
histos["PRED-GEN_Pt_Resol1"].Draw("SAMES")
st = root.TPaveStats()
st.FindObject("stats")
c8.Update()
leg8 = root.TLegend(0.49,0.5,0.90,0.63);
leg8.SetTextSize(0.04);
leg8.SetHeader("p_{T} Resolution:"); 
leg8.AddEntry(histos["PRED-GEN_Pt_Resol1"], "NEW: Machine Learning", "l");
leg8.AddEntry(histos["L1-GEN_Pt_Resol1"], "CURRENT: Level-1 trigger", "l");
leg8.Draw();
c8.Write("Pt-resol0-20")

c9 = root.TCanvas("c9","c9");
histos["L1-GEN_Pt_Resol2"].SetLineColor(kRed)
histos["L1-GEN_Pt_Resol2"].SetTitle("p_{T} resolution comparison - Range 20-40 GeV")
histos["PRED-GEN_Pt_Resol2"].SetLineColor(kBlue)
histos["L1-GEN_Pt_Resol2"].Draw()
histos["PRED-GEN_Pt_Resol2"].Draw("SAMES")
st = root.TPaveStats()
st.FindObject("stats")
c9.Update()
leg9 = root.TLegend(0.49,0.5,0.90,0.63);
leg9.SetTextSize(0.04);
leg9.SetHeader("p_{T} Resolution:"); 
leg9.AddEntry(histos["PRED-GEN_Pt_Resol2"], "NEW: Machine Learning", "l");
leg9.AddEntry(histos["L1-GEN_Pt_Resol2"], "CURRENT: Level-1 trigger", "l");
leg9.Draw();
c9.Write("Pt-resol20-40")

c10 = root.TCanvas("c10","c10");
histos["L1-GEN_Pt_Resol3"].SetLineColor(kRed)
histos["PRED-GEN_Pt_Resol3"].SetTitle("p_{T} resolution comparison - Range 40-60 GeV")
histos["PRED-GEN_Pt_Resol3"].SetLineColor(kBlue)
histos["PRED-GEN_Pt_Resol3"].Draw()
histos["L1-GEN_Pt_Resol3"].Draw("SAMES")
st = root.TPaveStats()
st.FindObject("stats")
c10.Update()
leg10 = root.TLegend(0.49,0.5,0.90,0.63);
leg10.SetTextSize(0.04);
leg10.SetHeader("p_{T} Resolution:"); 
leg10.AddEntry(histos["PRED-GEN_Pt_Resol3"], "NEW: Machine Learning", "l");
leg10.AddEntry(histos["L1-GEN_Pt_Resol3"], "CURRENT: Level-1 trigger", "l");
leg10.Draw();
c10.Write("Pt-resol40-60")

c11 = root.TCanvas("c11","c11");
histos["L1-GEN_Pt_Resol4"].SetLineColor(kRed)
histos["PRED-GEN_Pt_Resol4"].SetTitle("p_{T} resolution comparison - Range 60-80 GeV")
histos["PRED-GEN_Pt_Resol4"].SetLineColor(kBlue)
histos["PRED-GEN_Pt_Resol4"].Draw()
histos["L1-GEN_Pt_Resol4"].Draw("SAMES")
st = root.TPaveStats()
st.FindObject("stats")
c11.Update()
leg11 = root.TLegend(0.49,0.5,0.90,0.63);
leg11.SetTextSize(0.04);
leg11.SetHeader("p_{T} Resolution:"); 
leg11.AddEntry(histos["PRED-GEN_Pt_Resol4"], "NEW: Machine Learning", "l");
leg11.AddEntry(histos["L1-GEN_Pt_Resol4"], "CURRENT: Level-1 trigger", "l");
leg11.Draw();
c11.Write("Pt-resol60-80")

c12 = root.TCanvas("c12","c12");
histos["L1-GEN_Pt_Resol5"].SetLineColor(kRed)
histos["PRED-GEN_Pt_Resol5"].SetTitle("p_{T} resolution comparison - Range 80-100 GeV")
histos["PRED-GEN_Pt_Resol5"].SetLineColor(kBlue)
histos["PRED-GEN_Pt_Resol5"].Draw()
histos["L1-GEN_Pt_Resol5"].Draw("SAMES")
st = root.TPaveStats()
st.FindObject("stats")
c12.Update()
leg12 = root.TLegend(0.49,0.5,0.90,0.63);
leg12.SetTextSize(0.04);
leg12.SetHeader("p_{T} Resolution:"); 
leg12.AddEntry(histos["PRED-GEN_Pt_Resol5"], "NEW: Machine Learning", "l");
leg12.AddEntry(histos["L1-GEN_Pt_Resol5"], "CURRENT: Level-1 trigger", "l");
leg12.Draw();
c12.Write("Pt-resol80-100")

c13 = root.TCanvas("c13","c13");
histos["L1-GEN_Pt_Resol8"].SetLineColor(kRed)
histos["L1-GEN_Pt_Resol8"].SetTitle("p_{T} resolution comparison - Range 3-120 GeV")
histos["PRED-GEN_Pt_Resol8"].SetLineColor(kBlue)
histos["L1-GEN_Pt_Resol8"].Draw()
histos["PRED-GEN_Pt_Resol8"].Draw("SAMES")
st = root.TPaveStats()
st.FindObject("stats")
c13.Update()
leg13 = root.TLegend(0.49,0.5,0.90,0.63);
leg13.SetTextSize(0.04);
leg13.SetHeader("p_{T} Resolution:"); 
leg13.AddEntry(histos["PRED-GEN_Pt_Resol8"], "NEW: Machine Learning", "l");
leg13.AddEntry(histos["L1-GEN_Pt_Resol8"], "CURRENT: Level-1 trigger", "l");
leg13.Draw();
c13.Write("Pt-resol0-120")

print("Level-1 5-15 error: " + str((1/all_count)*(np.sqrt(count*(1-(count/all_count))))))
print("ML 5-15 error: " + str((1/all_count2)*(np.sqrt(count2*(1-(count2/all_count2))))))
print("Level-1 30-190 error: " + str((1/all_count3)*(np.sqrt(count3*(1-(count3/all_count3))))))
print("ML 30-190 error: " + str((1/all_count4)*(np.sqrt(count4*(1-(count4/all_count4))))))
print("The level-1 medium eff at 5-15 is " + str(count/all_count))
print("The ML medium eff at 5-15 is " + str(count2/all_count2))
print("The level-1 medium eff at 30-190 is " + str(count3/all_count3))
print("The ML medium eff at 30-190 is " + str(count4/all_count4))

outputFile.Write()
outputFile.Close()
bar.finish()
```
